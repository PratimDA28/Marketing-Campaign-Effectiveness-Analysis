```python
'''
Frequency → shows how often customers buy (behavior)
Tenure → separates new vs long-term customers

These two are mandatory for:
baseline understanding
responder vs non-responder analysis
segmentation later'''

# Purchase frequency
data["Frequency"] = (data["NumWebPurchases"] +data["NumCatalogPurchases"] +data["NumStorePurchases"])

# Customer tenure in years
data["Tenure_Years"] = data["Customer_Days"] / 365

data["Frequency"]
data["Tenure_Years"]
```

```python
'''Establish the baseline
Creates the reference point for all comparisons
Prevents misleading insights later
This is how analysts stay objective'''

# Total customers
total_customers = data.shape[0]

# Average & median spend
avg_spend = data["MntTotal"].mean()
median_spend = data["MntTotal"].median()

# Average income
avg_income = data["Income"].mean()

# Average purchase frequency
avg_frequency = data["Frequency"].mean()

# Campaign response rate
response_rate = data["Response"].mean()

total_customers, avg_spend, median_spend, avg_income, avg_frequency, response_rate
```

```python
data["MntTotal"]
```

```python
'''Revenue concentration (who drives the money)
Identify how much revenue comes from the top customers.
Shows whether revenue is concentrated or spread
Helps marketing focus on high-impact customers
This is a classic executive-level insight'''

# Sort customers by total spend
data_sorted = data.sort_values("MntTotal", ascending=False)

# Top 20% customers
top_20_pct = int(0.2 * len(data_sorted))
top_customers = data_sorted.head(top_20_pct)

# Revenue contribution
total_revenue = data_sorted["MntTotal"].sum()
top_revenue = top_customers["MntTotal"].sum()

top_revenue_share = top_revenue / total_revenue
top_revenue_share
```

Top 20% customers generate 53.6% of the total revenue.
Losing top customers would hurt revenue significantly.
Marketing should prioritize retention and upsell for this group

```python
'''Responder vs Non-Responder comparison
Split customers based on campaign response and compare core metrics.
Identifies who actually responds to marketing
Directly informs targeting strategy
This is the core client question'''

responders = data[data["Response"]==1]
non_responders = data[data["Response"] == 0]

comparison = {
    "Avg_Income": [
        responders["Income"].mean(),
        non_responders["Income"].mean()
    ],
    "Avg_Total_Spend": [
        responders["MntTotal"].mean(),
        non_responders["MntTotal"].mean()
    ],
    "Avg_Frequency": [
        responders["Frequency"].mean(),
        non_responders["Frequency"].mean()
    ],
    "Avg_Recency": [
        responders["Recency"].mean(),
        non_responders["Recency"].mean()
    ]
}

comparison
```

```python
'''Product-level revenue contribution
Which product categories drive revenue.
Tells what to promote in campaigns
Prevents wasting effort on low-impact products
Directly informs offer design'''

product_cols = ["MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", "MntSweetProducts", "MntGoldProds"]

product_revenue = data[product_cols].sum()
product_revenue

product_share = product_revenue / product_revenue.sum()
product_share

product_share.sort_values(ascending=False)

```

```python
'''
Channel preference analysis
Determine how customers prefer to purchase.
Tells where to run campaigns
Prevents channel misalignment
Needed for Power BI storytelling'''


data["Web_Share"] = data["NumWebPurchases"] / data["Frequency"]
data["Store_Share"] = data["NumStorePurchases"] / data["Frequency"]
data["Catalog_Share"] = data["NumCatalogPurchases"] / data["Frequency"]

channel_pref = data[["Web_Share", "Store_Share", "Catalog_Share"]].mean()
channel_pref
```

```python
'''
Simple customer segmentation
Create value-based segments using total spend
Marketing works with segments, not raw customers
This enables targeting, exclusion, and prioritization
Required for Power BI dashboards'''

# Create spend quantiles
high_value_threshold = data["MntTotal"].quantile(0.75)
low_value_threshold = data["MntTotal"].quantile(0.25)

def value_segment(spend):
    if spend >= high_value_threshold:
        return "High Value"
    elif spend <= low_value_threshold:
        return "Low Value"
    else:
        return "Mid Value"

data["Value_Segment"] = data["MntTotal"].apply(value_segment)

# Check segment distribution
data["Value_Segment"].value_counts(normalize=True)
```
